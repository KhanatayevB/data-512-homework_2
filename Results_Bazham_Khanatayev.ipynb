```python
import pandas as pd
```

```python
# Load the CSV file into a dataframe
df_input = pd.read_csv('wp_scored_city_articles_by_state.csv')
```

```python
df_input
```

 ## 1.Top 10 US states by coverage: The 10 US states with the highest total articles per capita (in descending order).

```python
# Assuming df_input is already loaded with the CSV data
df_input['articles_per_capita'] = df_input.groupby('state')['article_title'].transform('count') / df_input['population']
```

```python
# Sort states by articles per capita in descending order and drop duplicates
top_states_by_coverage = df_input.sort_values('articles_per_capita', ascending=False).drop_duplicates(subset='state').head(10)
```

```python
# Display the top 10 states in a nice table format
top_states_by_coverage = top_states_by_coverage[['state', 'articles_per_capita']]
```

```python
display(top_states_by_coverage.set_index('state'))
```

## 2. Bottom 10 US states by coverage: The 10 US states with the lowest total articles per capita (in ascending order) .

```python
# Sort states by articles per capita in ascending order and drop duplicates
bottom_states_by_coverage = df_input.sort_values('articles_per_capita').drop_duplicates(subset='state').head(10)
```

```python
# Display the bottom 10 states in a nice table format
bottom_states_by_coverage = bottom_states_by_coverage[['state', 'articles_per_capita']]
```

```python
display(bottom_states_by_coverage.set_index('state'))
```

Next, we need to define the order of the rankings. 

```python
# Get the unique quality scores
unique_quality_scores = df_input['article_quality'].unique()

unique_quality_scores
```

```python
# Define the ranking
quality_ranking = {
    'FA': 1,    # Featured article
    'GA': 2,    # Good article
    'B': 3,     # B-class article
    'C': 4,     # C-class article
    'Start': 5, # Start-class article
    'Stub': 6   # Stub-class article
}
```

```python
# Map the article_quality column to numeric values based on the ranking
df_input['quality_rank'] = df_input['article_quality'].map(quality_ranking)
```

```python
df_input
```

```python
df_input['quality_rank'].unique()
```

Now we can work on the next question.

## 3. Top 10 US states by high quality: The 10 US states with the highest high quality articles per capita (in descending order) .


```python
# Filter the dataframe for articles with quality rank 1 or 2
quality_1_2 = df_input[df_input['quality_rank'].isin([1, 2])].copy()
```

```python
# Group by state and get the count of quality rank 1 and 2 articles
quality_count_per_state = quality_1_2.groupby('state').size()
```

```python
quality_1_2
```

```python
# Calculate the ratio of quality rank 1 and 2 articles per capita for each state
articles_per_capita = quality_count_per_state / df_input.groupby('state')['population'].first()
```

```python
# Sort the states by this metric in descending order
sorted_states = articles_per_capita.sort_values(ascending=False).head(10)
```

```python
# Convert the series to a dataframe for better display
highest_hq_articles = sorted_states.reset_index()
highest_hq_articles.columns = ['State', 'Articles per Capita']
```

```python
highest_hq_articles
```

## 4. Bottom 10 US states by high quality: The 10 US states with the lowest high quality articles per capita (in ascending order).

```python
# Sort the states by this metric in ascending order and get the bottom 10
sorted_states_bottom = articles_per_capita.sort_values().head(10)
```

```python
# Convert the series to a dataframe for better display
lowest_hq_articles = sorted_states_bottom.reset_index()
lowest_hq_articles.columns = ['State', 'Articles per Capita']
```

```python
# Display the table in Jupyter Notebook
display(lowest_hq_articles.set_index('State'))
```

## 5. Census divisions by total coverage: A rank ordered list of US census divisions (in descending order) by total articles per capita.

```python
# Group by regional_division and calculate the total number of articles and total population
grouped_divisions = df_input.groupby('regional_division').agg({'article_title': 'count', 'population': 'sum'})
```

```python
# Calculate articles per capita for each division
grouped_divisions['articles_per_capita'] = grouped_divisions['article_title'] / grouped_divisions['population']
```

```python
# Sort the divisions by articles per capita in descending order
sorted_divisions = grouped_divisions.sort_values(by='articles_per_capita', ascending=False)
```

```python
# Display the sorted divisions with their articles per capita ratio in a table
table_divisions = sorted_divisions[['articles_per_capita']]
```

```python
display(table_divisions)
```

## 6. Census divisions by high quality coverage: Rank ordered list of US census divisions (in descending order) by high quality articles per capita.

```python
# Filter the dataframe for articles with quality rank 1 or 2
high_quality_articles = df_input[df_input['quality_rank'].isin([1, 2])]
```

```python
# Group by regional_division and calculate the total number of high-quality articles and total population
grouped_divisions_high_quality = high_quality_articles.groupby('regional_division').agg({'article_title': 'count', 'population': 'sum'})
```

```python
# Calculate high-quality articles per capita for each division
grouped_divisions_high_quality['high_quality_articles_per_capita'] = grouped_divisions_high_quality['article_title'] / grouped_divisions_high_quality['population']
```

```python
# Display the divisions with their high-quality articles per capita ratio in a table
table_high_quality_divisions = grouped_divisions_high_quality[['high_quality_articles_per_capita']]
```

```python
# Sort the divisions by high-quality articles per capita in descending order
sorted_high_quality_divisions = table_high_quality_divisions.sort_values(by='high_quality_articles_per_capita', ascending=False)
```

```python
# Display the sorted table
display(sorted_high_quality_divisions)
```
