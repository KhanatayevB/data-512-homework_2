Population Data Preparation

```python
df_pop_states = pd.read_excel("NST-EST2022-POP_Clean.xlsx", skiprows = 1)
```

We are going to take the top 5 rows and move them into a new df. The top 5 rows only show regional and country information and not state information. It is still useful to store this information for future comparisons.

```python
df_pop_regions = df_pop_states.iloc[:5].copy()
```

```python
df_pop_states = df_pop_states.iloc[5:].reset_index(drop=True)
```

```python
df_pop_regions
```

```python
df_pop_states
```

There is a period before every state name in the Geographic Area column. We need to remove that period.

```python
# Replace the period in the 'Geographic Area' column
df_pop_states['Geographic Area'] = df_pop_states['Geographic Area'].str.replace('.', '', regex=False)
```

Prepare Quality Scores

```python
df_quality_scores = pd.read_csv("quality_scores_pred.csv")
```

```python
# 1. Split Article Title column
df_quality_scores['City'] = df_quality_scores['Article Title'].str.rsplit(',', 1).str[0]
df_quality_scores['State'] = df_quality_scores['Article Title'].str.rsplit(',', 1).str[1].str.strip().fillna('')
```

Region Join

```python
# Copy df_quality_scores to df_region_join
df_region_join = df_quality_scores.copy()
```

```python
# Merge based on the state
df_region_join = df_region_join.merge(df_regions[['STATE', 'REGION', 'DIVISION']], 
                                      left_on='State', 
                                      right_on='STATE', 
                                      how='left')
```

```python
# Drop the STATE column as it's redundant
df_region_join.drop('STATE', axis=1, inplace=True)
```

Join Population

```python
# Copy df_region_join to df_region_pop_join
df_region_pop_join = df_region_join.copy()
```

```python
# Merge the dataframes based on the respective columns
df_region_pop_join = df_region_pop_join.merge(df_pop_states, 
                                             left_on='State', 
                                             right_on='Geographic Area', 
                                             how='left')
```

```python
# Drop the 'Geographic Area' column as it's redundant after the merge
df_region_pop_join.drop('Geographic Area', axis=1, inplace=True)
```

We are going to store the population rows that did not get matched to an article row.

```python
# Perform a merge with an indicator
merged = df_pop_states.merge(df_region_join, 
                             left_on='Geographic Area', 
                             right_on='State', 
                             how='left', 
                             indicator=True)
```

```python
# Filter out rows that didn't find a match in df_region_join
pop_not_merged = merged[merged['_merge'] == 'left_only']
```

```python
# Keep only the original columns from df_pop_states
pop_not_merged = pop_not_merged[df_pop_states.columns]
```
